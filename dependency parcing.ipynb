```python
os.environ["STANFORD_PARSER"] = "/afs/inf.ed.ac.uk/user/s16/s1678999/Documents/Msc project/StanfordParser/stanford-parser-full-2017-06-09"
os.environ["STANFORD_MODELS"] = "/afs/inf.ed.ac.uk/user/s16/s1678999/Documents/Msc project/StanfordParser/stanford-parser-full-2017-06-09"
os.environ["CLASSPATH"] = '/afs/inf.ed.ac.uk/user/s16/s1678999/Documents/Msc\ project/StanfordParser/stanford-ner-2017-06-09/stanford-ner.jar:/afs/inf.ed.ac.uk/user/s16/s1678999/Documents/Msc\ project/StanfordParser/stanford-parser-full-2017-06-09/stanford-parser.jar:/afs/inf.ed.ac.uk/user/s16/s1678999/Documents/Msc\ project/StanfordParser/stanford-postagger-2017-06-09/stanford-postagger.jar'
home_path = './stanford-parser-models/'
Cell_Phones_and_Accessories = pd.read_json('./data/reviews_Cell_Phones_and_Accessories_5.json', lines=True)
parser = StanfordDependencyParser(model_path=home_path+'edu/stanford/nlp/models/lexparser/englishPCFG.ser.gz')
Cell_Phones_and_Accessories['index'] = np.arange(len(Cell_Phones_and_Accessories))
```

```python
result_path = './dependencies/'
with tqdm(total=len(Cell_Phones_and_Accessories)) as pbar:
    for review,index in zip(Cell_Phones_and_Accessories.iloc[:,3],Cell_Phones_and_Accessories.iloc[:,9]):
        try:
            sentences = nltk.sent_tokenize(review)
            results = []
            for sentence in sentences:
                parses = parser.raw_parse(sentence)
                results += [list(parse.triples()) for parse in parses][0]
            with open(result_path+str(index),'w') as f:
                pickle.dump(results, f)
        except:
            print('error:'+str(index))
        finally:
            pbar.update(1)
```
